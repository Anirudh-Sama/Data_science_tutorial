```python
# Indexing the data using the [] operator
import pandas as pd

# Displaying the data
data = pd.read_csv("C:/Users/15145/Desktop/Data_science/datas/nba.csv", index_col = "Name")
print("Dataset")
display(data.head(10))
```

```python
# Displaying the data with a single column after the name index column
first = data["Age"]
print("\n Single column selected from dataset")
display(first.head(5))
```

```python
#Selecting multiple columns
second = data[["Age", "Salary", "College"]]
print(" \n Multiple Columns selected from the dataset")
display(second.head(10))
```

```python
# Indexing with .loc[]
```

```python
#Selecting a single row with label
row = data.loc["Avery Bradley"]
print(row)
print("\n")

# Selecting multiple rows with labels
rows = data.loc[["Avery Bradley", "R.J. Hunter"]]
print(rows)
print("\n")

# Selecting specific rows and columns
rows2 = data.loc[["Avery Bradley","R.J. Hunter"], ["Team", "Number", "Position"]]
print(rows2)
print("\n")

#Selecting specific columns and all rows
rows3 = data.loc[:, ["Team", "Position", "Salary"]]
print(rows3)
```

```python
#Indexing with iloc
```

```python
# Selecting a single row by position
row = data.iloc[3]
print(row)
print("\n")

# Selecting multiple rows by position
row1 = data.iloc[[3,5,7]]
print(row1)
print("\n")

#Selecting specific rows and columns by position
row2 = data.iloc[[3, 4], [1, 2]]
print(row2)
print("\n")

#Selecting all rows and specific columns
row3 = data.iloc[: , [1, 2]]
print(row3)
```

```python
#Other Useful methods
```

```python
# .head()
print(data.head(2))
print("\n")

# .tail()
print(data.tail(2))
print("\n")

# .at()
value = data.at["Avery Bradley", "Age"]
print("Avery's age is", value)
print("\n")

# .query()
result = data.query("Age > 25 and College == 'Duke'")
print(result)
```
