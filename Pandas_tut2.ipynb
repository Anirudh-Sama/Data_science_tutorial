```python
# Indexing the data using the [] operator
import pandas as pd

# Displaying the data
data = pd.read_csv("C:/Users/15145/Desktop/Data_science/datas/nba.csv", index_col = "Name")
print("Dataset")
display(data.head(10))
```

```python
# Displaying the data with a single column after the name index column
first = data["Age"]
print("\n Single column selected from dataset")
display(first.head(5))
```

```python
#Selecting multiple columns
second = data[["Age", "Salary", "College"]]
print(" \n Multiple Columns selected from the dataset")
display(second.head(10))
```

```python
# Indexing with .loc[]
```

```python
#Selecting a single row with label
row = data.loc["Avery Bradley"]
print(row)
print("\n")

# Selecting multiple rows with labels
rows = data.loc[["Avery Bradley", "R.J. Hunter"]]
print(rows)
print("\n")

# Selecting specific rows and columns
rows2 = data.loc[["Avery Bradley","R.J. Hunter"], ["Team", "Number", "Position"]]
print(rows2)
```
